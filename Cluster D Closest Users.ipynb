```python
import codecs
import numpy as np
from pprint import pprint
from numpy.linalg import norm
```

```python
# process centroids
# infile = "/Users/curtisxuan/Desktop/Work/KNN_Users/centroids.txt"

infile = "/Users/curtisxuan/Desktop/Work/experiment2/centroidsD.txt"
centroids = []
with codecs.open(infile, "r", "utf-8") as f:
    for line in f:
        dim = line.rstrip().split(" ")
        centroids.append(dim)
centroids = np.array (centroids,dtype='float32')
centroids = centroids[:,1:]
# print(centroids[31])


usersVec = "/Users/curtisxuan/Desktop/Work/experiment2/userVecD.txt"
ids=[]
values=[]
count=0
with codecs.open(usersVec, 'r', 'utf8') as f:
    for line in f:
        user = line.rstrip().split(' ')
        temp = np.array (user[1:],dtype='float32')
        L2norm = norm(temp, ord=2)
        if L2norm>0:
            ids.append(user[0])
            values.append(user[1:])
        count+=1
print(len(ids))
print(len(values))
print(count)

X = np.array (values,dtype='float32')
# #normalize data, Y is normalized X
L2norm = norm(X, axis=1, ord=2)
Y=X / L2norm.reshape(len(ids),1)
print('done')
```

```python
#find top 10 closest user to each centroid
print('run')
ct=1
minList=[]
for centroid in centroids:
    tempRaw = Y
    tempFea = ids
    top10 = []
#     print(centroid)
#     print(' ')
    dist=[]
    for feature in tempRaw:
            dist.append(np.linalg.norm(centroid-feature))
    for iteration in range(1,11):
        count=0
        minDist=100
        minIndex=0
        for distance in dist:
            if minDist > distance:
                minDist = distance
                minIndex = count
            count+=1
#         print('minINdex is ' + str(minIndex))
#         print(tempFea[minIndex]) #user id
#         print(tempRaw[minIndex]) #vector
#         print(minDist)
        top10.append([iteration,tempFea[minIndex],minDist]) #iteration, user id, distance
        tempRaw=np.delete(tempRaw,minIndex,0) #remove vector 
        tempFea.pop(minIndex) #remove user id
        dist.pop(minIndex)
    print ([ct,top10])
    minList.append([ct,top10])
    ct+=1
pprint(minList)

```
